Main OCR function

```python
def main_ocr(current_page, color, black):
    """
    Inputs : 
    scanned page
    color of the page
    
    Output : 
    text output from image_to_string method
    
    """
    
    # step 1 
    current_page.save('temp1.jpg', 'JPEG') # save current page as JPEG
    
    
    # step 2
    if color == 'g': # color parameter
        read_image = cv2.imread("temp1.jpg", cv2.IMREAD_COLOR)     # image read
        gray_image = cv2.cvtColor(read_image, cv2.COLOR_BGR2GRAY)  # turn image to gray.
        cv2.imwrite("temp2.jpg", gray_image)               # save as gray
        
    elif color == 'b':
        read_image = cv2.imread("temp1.jpg", cv2.IMREAD_COLOR)     # image read
        gray_image = cv2.cvtColor(read_image, cv2.COLOR_BGR2GRAY)  # turn image to gray.
        (thresh, BnW_image) = cv2.threshold(gray_image, black, 255, cv2.THRESH_BINARY) # this controls color temp
        cv2.imwrite("temp2.jpg", BnW_image)                # save as black and white

    
    # step 3
    text = pytesseract.image_to_string('temp2.jpg', lang='ben') # scan page

    return text
```

Total number of pmages read

```python
total_page = len(pages)
total_page
```

Main Data Processing Section

```python
# test complete? if yes then continue

start_page = 0          # first page
end_page = total_page   # last page

dataList = []   # storage of each page as a list

for i in range(start_page, end_page):
    print("Scaning page:", i+1)
    dataList.append( main_ocr(pages[i], 'g', 160) )
```

### Paer 2: Clean Data 

#### Split each page using `pattern='\n'`

```python
data = dataList.copy()

for i, _data in enumerate(dataList):
    #data[i] =_data.split('\n')             # i index & data
    data[i] =data[i].split('\n')             # split where "\n"
```

The pages of the book now appears as a list of lists.

Clean text

Search for patterns in the text. When found, just remove.

```python
for i, _data in enumerate(data):
    temp = [] # temp store
    for line in _data:
        temp.extend(line.split("|")) # where string will be split
    data[i] = temp

for i, _data in enumerate(data):
    temp = []
    for line in _data:
        temp.extend(line.split(" ")) # where string will be split
    data[i] = temp
    
for i, _data in enumerate(data):
    temp = []
    for line in _data:
        temp.extend(line.split("-")) # where string will be split
    data[i] = temp

for i, _data in enumerate(data):
    temp = []
    for line in _data:
        temp.extend(line.split(",")) # where string will be split
    data[i] = temp
    
for i, _data in enumerate(data):
    temp = []
    for line in _data:
        temp.extend(line.split("?")) # where string will be split
    data[i] = temp
    
for i, _data in enumerate(data):
    temp = []
    for line in _data:
        temp.extend(line.split("!")) # where string will be split
    data[i] = temp
```

Further clean text

Remove special characters

```python
# remove or ignore all chars present in the 'ignore_list'
ignore_list = "!#$%&'()\n”*+”“,-.৷'/:;<=>?[]@\^_|`{|}~।ঃ"+'"' 

for i, _data in enumerate(data):
    temp = []
    
    for word in _data:
#         print(word)

        for char in word:
            if char in ignore_list:
                word = word.replace(char, '')
            if char.isdigit():
                word = word.replace(char, '') 
        if len(word) != 0: 
            temp.append(word)    
    data[i] = temp
```

Create a dictionary : each page numbger is the key, the list of words in each page is the value

```python
dictionary = {}

for i ,_data in enumerate(data):
    dictionary[i] = _data
```

#### Do you see the potential challenges in extarcting Bangla text using pytesseract?

How else would you proceed? in terms of data, in terms of algorithm.
The  challenges in extracting Bangla text using Pytesseract are:
1. Pytesseract doesn't have the Bengali language by default.
2. Recognition of complex Bangla characters is difficult if they are not in ideal font size, style, or different orientations.
As a result, the OCR result of the processed image may not be as good as the original one.

In terms of data, we need ideal PDF/image quality, font size, and style for proper text recognization. After data cleaning, some words are still misspelled.
So, we need a spellchecker for correcting misspelled words. Also, we can improve our algorithm so that it can recognize properly any type of font size, and style, at any angle and correct the spelling error,typo.
